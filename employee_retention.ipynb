```python
# Convert the 'salary' column into numeric values
df1=df.copy()
df['salary_encoded'] = df['salary'].map({'low': 0, 'medium': 1, 'high': 2})

df = df.drop('salary', axis=1)
## Checking for correlation
cor_mat=df.corr()
fig = plt.figure(figsize=(15,7))
sns.heatmap(cor_mat,annot=True)
```

```python
df2=df1.copy()
df2.drop(['empid'],axis=1,inplace=True)
```

```python
df2.shape
```

```python
#handle missing values
df2.isnull().sum()
```

```python
df2['satisfaction_level'].fillna(df2['satisfaction_level'].mean(), inplace=True)
```

```python
df2.isnull().sum()
```

```python
#handle categorical features..salary using get dummies
df2['salary'].unique()
```

```python
salary_dummies = pd.get_dummies(df2['salary'],drop_first=True)
salary_dummies
```

```python
df2=pd.concat([df2,salary_dummies],axis=1)
```

```python
#drop salary feature as we already applied lable endcoding tech
df2.drop(['salary'],axis=1,inplace=True)
```

## Split dataset into training set and test set

```python
#split datset into train and target
x=df2.drop(['left'],axis=1)
y=df2['left']
#split dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

### Logistic regression Model 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict_proba(x_test)
y_pred
```

```python

y_pred = model.predict_proba(x_test)[:, 1]
left = y_pred >= 0.4
accuracy=(y_test == left).mean()
accuracy
```
